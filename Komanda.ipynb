```python
import pandas as pd

# Видаляємо зайві пробіли у назвах колонок
df.columns = df.columns.str.strip()

# Виводимо перші 5 рядків
print("Перші 5 рядків набору даних:")
display(df.head())

# Перевіряємо загальну інформацію про дані
print("\nЗагальна інформація про набір даних:")
df.info()

# Описуємо числові характеристики
print("\nСтатистичний опис числових змінних:")
display(df.describe())

# Описуємо категоріальні змінні
print("\nУнікальні значення категоріальних змінних:")
for col in df.select_dtypes(include=['object']).columns:
    print(f"{col}: {df[col].nunique()} унікальних значень")
    print(df[col].value_counts().head(), "\n")

# Аналіз цільової змінної (Churn)
target_col = 'churn' if 'churn' in df.columns else 'Churn'  # Перевіряємо різні варіанти назви
print("\nАналіз цільової змінної (Churn):")
display(df[target_col].value_counts(normalize=True) * 100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Очистка назв колонок
df.columns = df.columns.str.strip()

# Перевірка пропущених значень
print("\nАналіз пропущених значень:")
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print(missing_values)

# Візуалізація пропущених значень
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title("Карта пропущених значень")
plt.show()

# Гістограми для числових змінних
df.hist(figsize=(12, 8), bins=30, edgecolor="black")
plt.suptitle("Розподіли числових змінних", fontsize=16)
plt.show()

# Аналіз кореляції
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Кореляція між змінними")
plt.show()

# Аналіз цільової змінної (Churn)
target_col = 'churn' if 'churn' in df.columns else 'Churn'
plt.figure(figsize=(5, 4))
sns.countplot(x=df[target_col], palette="coolwarm")
plt.title("Розподіл Churn")
plt.xlabel("Churn")
plt.ylabel("Кількість клієнтів")
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Очистка назв колонок
df.columns = df.columns.str.strip()

# 1Заповнення пропущених значень
print("\nАналіз пропущених значень перед обробкою:")
print(df.isnull().sum())

# Заповнення числових змінних середнім значенням
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Кодування категоріальних змінних
cat_cols = df.select_dtypes(include=['object']).columns

if len(cat_cols) > 0:
    print("\nКатегоріальні змінні, які будуть закодовані:", cat_cols.tolist())

    # Використовуємо Label Encoding для змінних з 2 категоріями, інакше One-Hot Encoding
    for col in cat_cols:
        if df[col].nunique() == 2:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col])
        else:
            df = pd.get_dummies(df, columns=[col], drop_first=True)

# Нормалізація числових ознак
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Вивід результатів
print("\nОбробка завершена! Фінальні дані:")
print(df.head())
```
